```python
def gerardados(nome_cliente, estados, produtos):

    vendas = []
        
# Gera dados para número do pedido e código do cliente
    for dados in range(1,101):

        nrpedido = dados
        codcliente = random.randint(1, 100)  
        cliente = random.choice(nome_cliente)      
        qtd =  random.randint(1, 100)
        produto = random.choice(produtos)
        custo = float(random.uniform(1, 100))
        valor_venda = float(custo * 1.5)
        data_atual = date.today()
        data = data_atual - timedelta(days=random.randint(1, 30))        
        uf = random.choice(estados)        
        vendas.append((nrpedido, codcliente, cliente, qtd, produto, custo, valor_venda, data, uf))

    return vendas  
```

```python
gerardados(nome_cliente, estados, produtos)
```

```python
spark = SparkSession.builder.appName(name="ecommerce").getOrCreate()
```

```python
# Definindo o schema

schema = StructType([
    StructField("nrPedido", IntegerType(), True),
    StructField("codCliente", IntegerType(), True),
    StructField("cliente", StringType(), True),
    StructField("qtd", IntegerType(), True),
    StructField("produto", StringType(), True),
    StructField("custo", FloatType(), True),
    StructField("valorVenda", FloatType(), True),
    StructField("data", DateType(), True),    
    StructField("uf", StringType(), True)    
])
```

```python
# Gerando dados fictícios de clientes para o DF
dados_vendas = gerardados(nome_cliente, estados, produtos)
```

```python
# Atribuindo os dados ao df
df = spark.createDataFrame(dados_vendas, schema)
```

# Trabalhando com datas

```python
# Ajustando a formatação da data para PT-BR
df.select(df.cliente, df.qtd, df.valorVenda, F.date_format('Data', "dd/MM/yyyy").alias("data")).show()
```

```python
# Consultando vendas em um período específico, sem utilizar o alias
df4 = df.select(
    df.produto,
    df.data.between("2024-06-01", "2024-06-10")).limit(10).show()
```

```python
# Consultando vendas em um período específico, utilizando o alias, para melhorar a visualização
df.select(
    df.produto,
    df.data,
    df.data.between("2024-06-01", "2024-06-10").alias("dentroPeriodo")).limit(10).show()
```

```python
# 5. *Custo Médio por Estado:  - Qual é o custo médio dos produtos vendidos em cada estado?
df.select(round(F.avg(col("custo")),2).alias("custoMedio")).show()
```

```python
# 6. *Total de Vendas por Cliente: Qual é o total de vendas (valor de venda) por cliente?

df_total_por_clientes = (
    df
    .groupBy("cliente")
    .agg(F.round(F.sum("valorVenda"),2).alias("totalPorCliente"))
    .withColumnRenamed("sum(valorVenda)", "totalPorCliente")
    .show(10)
)
```

```python
# 7. Data Mais Recente e Mais Antiga: Qual é a data mais recente e a mais antiga registrada nos dados?
df_antigo_recente = (
    df
    .select(min(df.data)
    .alias("antigo"),
    max(df.data)
    .alias("recente"))
    .show()
)
```

```python
# 8. *Quantidade de Vendas por Dia: Quantas vendas ocorreram em cada dia?
df_total_diario = (
    df
    .groupBy(date_format("data", "dd/MM/yyyy").alias("data"))
    .agg(round(sum("valorVenda"),2)
    .alias("TotalVendaDiaria"))
    .limit(10)
    .show()
)
```

```python
# 9. *Custo Médio e Valor de Venda Médio por Produto: Qual é o custo médio e o valor de venda médio para cada produto?
df_custoMedio_vendaMedia = (
    df.groupBy("produto")    
    .agg(round(avg("custo"),2)    
    .alias("custoMedio"),
    round(avg("valorVenda"))
    .alias("receitaMedia"))
    .limit(10)
    .show()
)
```

```python
# 10. *Porcentagem de Lucro: Qual é a porcentagem média de lucro (baseado no custo e valor de venda) em todas as vendas?
custo_venda = df.withColumn("custoTotal", (df.qtd * df.custo))
custo_venda.show()
```

```python
lucro_percentual = custo_venda.withColumn("lucroPercentual", (custo_venda.custoTotal * custo_venda.valorVenda) / 100)
lucro_percentual.show()
```

```python
#11) Data Mais Recente e Mais Antiga: Objetivo: Descobrir qual é a data mais recente e a mais antiga registrada nos dados.

data_recente_antiga = (
    df.select(
        F.max(date_format("data", "dd/MM/yyyy"))
        .alias("dataAntiga"), 
        F.min(date_format("data", "dd/MM/yyyy"))
        .alias("dataRecente"))
)

data_recente_antiga.show()
```

```python
# 12) Quantidade de Vendas por Dia. Objetivo: Contar quantas vendas ocorreram em cada dia.
#df.groupBy("data").agg(sum("qtd").alias("qtdDia")).show()
df_qtdDia = (
    df.groupBy("data")
    .agg(sum("qtd").alias("qtdDia"))    
    .sort("data")
)

df_qtdDia.show()
```

# Adicionar Coluna com Ano e Mês. Objetivo: Adicionar uma coluna que contenha apenas o ano e o mês da data da venda.

```python
# 13) Adicionar Coluna com Ano e Mês Objetivo: Adicionar uma coluna que contenha apenas o ano e o mês da data da venda.
df_ano_mes = df.withColumn("ano_mes", date_format(df.data, "yyyy-MM"))
df_ano_mes.show()
```

```python
# 14) Filtrar Vendas de um Mês Específico. Objetivo: Filtrar todas as vendas que ocorreram no mês de julho de 2024.
df.filter((year(col("data")) == 2024) & (month(col("data")) == 6)).show(10)
```

# Adicionar Coluna com Dia da Semana. Objetivo: Adicionar uma coluna que mostre o dia da semana para cada data de venda.

```python
# 15) Adicionar Coluna com Dia da Semana. Objetivo: Adicionar uma coluna que mostre o dia da semana para cada data de venda.
df_dia_semana = df.withColumn("dia_semana", date_format(df.data, "EEEE"))
df_dia_semana.show()
```

```python
# 16) Determinar o Dia da Semana com Mais Vendas
# Objetivo: Descobrir qual dia da semana teve o maior número de vendas totais (soma das quantidades vendidas).

df_dia_semana_total_qtd = (
    df_dia_semana
    .groupBy("dia_semana")
    .agg(sum("qtd").alias("qtd"))
    .sort("qtd", ascending=False)    
)

df_dia_semana_total_qtd.show()
```

```python
# 17) Calcular a Média de Vendas por Mês:
# Objetivo: Calcular a média da quantidade de vendas realizadas em cada mês.

df_media_vendas_mes = (
    df.groupBy(month("data")
               .alias("MesCorrespondente"))
    .agg(round(sum("valorVenda"),2)
         .alias("TotalPorMes"))
)

df_media_vendas_mes.show()
```
